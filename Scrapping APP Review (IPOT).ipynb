## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google_play_scraper import Sort, reviews
```

## Scraping Data Review

```python
result, continuation_token = reviews("com.indopremier.ipot",             # Alamat apps
                                     lang="id",                     # Bahasa Reviews , defaults : "en"
                                     country="id",                  # Negara Asal Reviewer defaults;  "us"
                                     sort=Sort.MOST_RELEVANT,       # Urutan Review yang diambil , 
                                     count=200)       
```

## Menyimpan Data Hasil Scrapping dalam Pandas Dataframe

```python
Hasil_Scrapping = pd.DataFrame(result)
Hasil_Scrapping
```

## Menyimpan Data Hasil Scrapping dalam file CSV

```python
Hasil_Scrapping.to_csv('Scrapping_IPOT.csv', index = False)
```
